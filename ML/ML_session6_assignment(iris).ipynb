```python
import pandas as pd
```

```python
link = r"C:\Users\dell\Documents\pincore\iris.csv"
ds=pd.read_csv(link)
print(ds.head(10))
```

```python
ds.info()
```

```python
x=ds.iloc[:,0:-1].values
y=ds.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=0)
```

```python
print("x_train-> ",x_train)
print("x_test-> ",x_test)
print("y_train-> ",y_train)
print("y_test-> ",y_test)
```

```python
#WITHOUT PREPROCESSING
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0) # vector can be for linear or no linear. Now lets assume that its linear but if non linear later we can change
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
print(y_predict,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuaracy level is : ',accuracy_score(y_test,y_predict)*100,"%")
print("Accuaracy level is: ",confusion_matrix(y_test,y_predict))
```

```python
#WITH PREPROCESSING
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
print("train ",x_train)
print("test ",x_test)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0) 
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
print(y_predict,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuaracy level is : ',accuracy_score(y_test,y_predict)*100,"%")
print("Accuaracy level is: ",confusion_matrix(y_test,y_predict))
```
