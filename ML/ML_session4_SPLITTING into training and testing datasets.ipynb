```python
x=ds.iloc[:,1:2].values #input features
y=ds.iloc[:,-1].values #output values
print(x)
print(y)
```

```python
#sklearn library 
#it had a module model_selection which has a package train_test_split to split as name says
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=0) 
# 4 variablesto split x n y as test and train datasets and size: 0.2 means 20% of the data(in the given datasets its has 200 rows) will be placed in the test datasets
#its better to give less test datsets and more training data sets so machines leans more 
#random_state to randomly pick the value 0 means it take 0.1 random data
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
print(y_train)
```

```python
# for linear regression
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
```

```python
#Fit function to fit the data for the linear regrssion plot
rg.fit(x_train,y_train) #Regression model to train machine using training data
```

```python
#Now model created so machine learnt the data 
#now we should test on the testing datasets using predict function of linear Regression class
y_predict=rg.predict(x_test)
```

```python
print(y_predict,y_test)
```

```python
#some are nearly same but some are bad so lets check the accuracy using score function of linearEegression class
accuracy=rg.score(x,y)
print("The accuracy level for RG is ",accuracy)
```

```python
#Now to get the plot of the regression model
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test,color='red') #test data plot
plt.plot(x_test,y_predict,color='blue') #predicated op for the testing data
plt.title("Sales prediaction based on TV advertisement")
plt.xlabel("Money spent")
plt.ylabel("Sales")
plt.show()
```
