```python
#plot the data grapghically
import matplotlib.pyplot as plt # python plot : to plot the result in python 
#kind: what kind of plot bar or line, box etc subplot: if true then it takes aech attribute as x vs y : if true then layout is number of axis
ds.plot(subplots=True,layout=(2,2))
plt.show()
```

```python
ds.plot(kind="box",subplots=True,layout=(2,2))
plt.show()
```

```python
#UNIVARIATE VISUALIZATION
ds.plot(subplots=True,layout=(2,2),sharex=False,sharey=False) # share y and x values 
plt.show()
```

```python
#BOX
ds.plot(kind="box",subplots=True,layout=(2,2),sharex=False,sharey=False) # share y and x values 
plt.show()
```
